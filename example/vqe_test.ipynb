```python
from isq import isq
import tensorflow as tf
```

```python
@isq.qpu(backend="simulate", shots=100)
#E_N=2,len(thetas)=6
def test(thetas,e_n):
    '''
    qbit q[4];
    X<q[0,1]>;


    Rx(1.57)<q[0]>;
    H<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rz(thetas[0])<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    H<q[3]>;
    Rx(1.57)<q[0]>;

    H<q[0]>;
    H<q[1]>;
    H<q[2]>;
    Rx(1.57)<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rz(thetas[1])<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rx(1.57)<q[3]>;
    H<q[2]>;
    H<q[1]>;
    H<q[0]>;


    Rx(1.57)<q[0]>;
    H<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rz(thetas[2])<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    H<q[3]>;
    Rx(1.57)<q[0]>;

    H<q[0]>;
    H<q[1]>;
    H<q[2]>;
    Rx(1.57)<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rz(thetas[3])<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rx(1.57)<q[3]>;
    H<q[2]>;
    H<q[1]>;
    H<q[0]>;


    Rx(1.57)<q[0]>;
    H<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rz(thetas[4])<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    H<q[3]>;
    Rx(1.57)<q[0]>;

    H<q[0]>;
    H<q[1]>;
    H<q[2]>;
    Rx(1.57)<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rz(thetas[5])<q[3]>;
    for i in 0:3 {
        CNOT<q[i],q[3]>;
    }
    Rx(1.57)<q[3]>;
    H<q[2]>;
    H<q[1]>;
    H<q[0]>;


    if (e_n == 0){
        Rx(1.57)<q[0]>;
        Rx(1.57)<q[1]>;
        H<q[2]>;
        H<q[3]>;
    }
    if (e_n == 1){
        Rx(1.57)<q[0]>;
        Rx(1.57)<q[1]>;
        H<q[2]>;
        H<q[3]>;
    }
    M<q[0:4]> ;
    '''
```

```python
def get_exception(thetas : list,E_N : int)->float :
    '''
    thetas制备时角度
    e_n<= E_N，测量第n个能量e
    '''
    exceptions = list()
    for e_n in range(E_N) :
        test_res=test(thetas,e_n)
        exception = 0
        for measure_res in test_res :
            frequency = test_res[measure_res]/100
            #频率代替概率
            parity = (-1)**measure_res.count('1')%2
            #奇偶效验
            exception += parity*frequency
        exceptions.append(exception)
    return sum(exceptions)
```

```python
gd_trains = tf.keras.optimizers.SGD(learning_rate=0.1)
for thetas in range :
    gd_trains.minimize(loss=get_exception,var_list=[thetas])
```
